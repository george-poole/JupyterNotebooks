## Two-State Approximation

By Bloch's theorem, the wavefunction of an electron in a one-dimensional periodic lattice must take the form

$$\psi(x)=u(x)e^{ikx}$$

where $u$ is a periodic function such that $u(x+a)=u(x)$. Expanding this as a Fourier series, we have

$$\psi(x)=\sum_{n=-\infty}^{\infty}C_n\phi_n(x)$$

with basis states $\phi_n(x)=e^{i(k+nG)x}/\sqrt{L}$ normalised on an interval length $L$ and $G=2\pi/a$. 

It is most convenient to tackle the Schrodinger equation by working with matrix elements. Subsituting our Fourier series into

$$\hat{H}\psi(x)=\epsilon\psi(x)$$

and using the orthonormality of basis states gives

$$\sum_nC_n\int_0^L\phi_m(x)\hat{H}\phi_n(x)=\epsilon C_m$$

in which the integral can be identified as the matrix element $H_{mn}$. Hence

$$\sum_nH_{mn}C_n~\text{d}x=\epsilon C_m$$

is an eigenvalue problem 

$$\textbf{H}\textbf{C}=\epsilon\textbf{C}$$

which is solved via the characteristic equation

$$\det(\textbf{H}-\epsilon\textbf{I})=0~~~.$$

The Hamiltonian is 

$$\hat{H}=\frac{1}{2m_e}\frac{\partial^2}{\partial x^2}+V(x)$$

where the potential $V(x)$ must be periodic. By symmetry it must also be symmetric about $x=0$ so its Fourier series is

$$V(x)=\sum_{j=1}^\infty V^{(j)}\cos(jGx)$$

where we have taken $V_0=0$ since the $j=0$ term only contributes an unimportant constant. Using the orthogonality conditions of the exponential functions gives matrix elements

$$\begin{align*}
H_{mn}&=\frac{\hbar(k+mG)^2}{2m_e}\delta_{mn}+\sum_jV^{(j)}(\delta_{m,n+j}+V_{m,n-j})\\
&=E^{(m)}\delta_{mn}+\sum_jV^{(j)}(\delta_{m,n+j}+V_{m,n-j})
\end{align*}$$


where a bracket in the superscript is used to distinguisg scalar functions from vector components.

In the two-state approximation, we need only consider the $n=0,-1$ terms so 

$$\psi(x)=C_0\phi_0(x)+C_{-1}\phi_{-1}(x)$$

so $\textbf{H}$ is a simple $2\times2$ matrix

$$\textbf{H}=\begin{pmatrix}
E_{-1} & V_1/2\\
V_1/2 & E_0
\end{pmatrix}$$

with distinct eigenvalues

$$\epsilon_\pm(k)=\tfrac{1}{2}(E^{(-1)}(k)+E^{(0)(k)})\pm\tfrac{1}{2}\sqrt{(E^{(-1)}(k)-E^{(0)}(k))^2+V_1^2}~~~.$$

Plots of $\epsilon_\pm(k)$ reveal two distinct curves which approach each other as $k\to G/2$, giving rise to a band gap structure. Series expansions can be used to show that the band gap is

$$\epsilon_+(G/2)-\epsilon_-(G/2)\approx|V_1|~~~.$$

Since we have symmetry under translations $x\to x+na$ in real space we must also have symmetry under translations $k\to k+nG$ in reciprocal space. Hence the rest of the energy landscape $\epsilon(k)$ outside the interval $[0,G]$ follows by symmetry.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

hbar=1
m=1
a=1
G=2*np.pi/a
V_1=1

def E(k,n):
    return hbar**2*(k+n*G)**2/2*m

def epsilon2(k,sign):
    return 0.5*(E(k,0)+E(k,-1))+sign*0.5*np.sqrt((E(k,0)-E(k,-1))**2+V_1**2)

plt.rc('text',usetex=True)
plt.rc('font',family='serif')

k=np.linspace(0,G,num=1000)
plt.plot(k/G,epsilon2(k,1),lw=0.5,color='b')
plt.plot(k/G,epsilon2(k,-1),lw=0.5,color='b')
plt.xlabel('$k/G$')
plt.ylabel('$\epsilon$')

plt.show()
```

## Beyond The Two-State Approximation

We can make a better approximation to the wavefunction with

$$\psi(x)=C_{-M}\phi_{-N}(x)+\dots C_N\phi_N(x)$$

where $N=M-1,M$ for an even or odd number of basis states respectively.

Matrix elements of the Hamiltonian are given by

$$H_{mn}=E^{(m)}\delta_{mn}+\sum_jV^{(j)}(\delta_{m,n+j}+V_{m,n-j})$$

with $m,n\in\{-M,\dots,N\}$ and evaluating these gives the general matrix

$$\textbf{H}=\begin{pmatrix}
E^{(-M)} & V_1/2 & \dots & & V_{N+M}/2\\
V_1/2 & E^{(-M+1)} &  &  & \\
\vdots &  & \ddots & & \vdots\\
 &  & & E^{(N-1)} & V_1/2\\
V_{N+M}/2 &  & \dots & V_1/2 & E^{(N)}\\
\end{pmatrix}$$

which has eigenvalues $\epsilon(k)$ that determine the band structure. 

If positive metal ions sit at the sites $x=na$, it makes sense for the electrons to experience a potential attracting them to $x=na$. Suppose that

$$V(x)=-\frac{1}{x-na}+\frac{1}{x-(n+1)a}$$

in the interval $[na,(n+1)a]$. Evaluating the Fourer series explicitly gives Fourier coefficients

$$V^{(j)}=...$$

Alternatively, the Fourier coefficients can simply be obtained numerically.

```python
def V(x,n):
    return -1/(x-n*a)+1/(x-(n+1)*a)

for n in range(-2,2):
    x=np.linspace(n*a+0.01,(n+1)*a-0.01)
    plt.plot(x,V(x,n),color='b')
    plt.plot(n*a,0,color='r',marker='o')
    
plt.plot(2*a,0,color='r',marker='o')
plt.xlabel('$x$')
plt.ylabel('$V(x)$')

plt.show()
```

```python
M=3

# dimensions of the matrix
D=2*M

# Fourier coefficients of the potential
V=np.array([0,4,-8,4,8,4,8,4,8,4,8,4,8,4,8,4,8,4,8,4,8,4,8,4,8])

def H(k,m,n):
    if m==n:
        return E(k,m)
    else:
        for j in range(1,len(V)):
            if j==m-n:
                return 0.5*V[j]
            if j==n-m:
                return 0.5*V[j]

def H_matrix(k):
    v=np.array([[0.0]*D]*D)
    for m in range(-M,D-M):
        for n in range(-M,D-M):
            v[m+M][n+M]=H(k,m,n)
    return v

def eigenvalue(k,index):
    return sorted(np.linalg.eig(H_matrix(k))[0])[index]
    
def epsilon(kval,index):
    v=[]
    for x in kval:
        v.append(eigenvalue(x,index))
    return v

k_array=np.linspace(0,G,num=1000)

for i in range(0,D):
    if i%2==0:
        plt.plot(k_array/G,epsilon(k_array,i),color='b')
    else:
        plt.plot(k_array/G,epsilon(k_array,i),color='r')
    
plt.xlabel('$k/G$')
plt.ylabel('$\epsilon$')
plt.show()
```

It can be shown that the $n^{\text{th}}$ band gap is, to a good approximation, the absolute value $|V_n|$ of coefficient in the potential's Fourier series. To demonstrate this mathematically for anything beyond the two state approximation would be incredibly tedious, but a computation can readily verify this.

```python
gap=[0.0]*(D-1)

for i in range(0,D-1):
    gap[i]=min(epsilon(k_array,i+1))-max(epsilon(k_array,i))

plt.plot(gap,color='g')
plt.plot(abs(V[1:len(gap)+1]),color='r')
plt.legend(['numerical','exact'])
plt.xlabel('$n$')
plt.ylabel('$|V_n|$')

plt.show()
```
