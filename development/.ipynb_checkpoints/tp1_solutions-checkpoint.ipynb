**(iii)** The degeneracy of the state with magnetisation $m$ is found by counting the possible ways of arranging $N$ objects, but there are $N+$ indistinguishable spin-up objects and $N_-$ indistinguishable spin-down objects.

$$W=\frac{N!}{N_+!N_-!}$$

Since $N=N_++N_-$ and $mN=\sum_is_i=N_+-N_-$, we can replace substitute for $N_\pm$ in the above expression.

**(iv)** The function $$F(m)=-Jm^2+k_BT\bigg((1+m)\ln(1+m)+(1-m)\ln(1-m)\bigg)$$
is only defined in the interval $-1<m<1$ and is since $F(m)=F(-m)$ it is also symmetric about $m=0$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

J=1
k_B=1
T_c=J/k_B
T=np.array([0.3,0,-0.3])
T=T+T_c

def F(x,i):
    return -J*x**2 + k_B*T[i]*(1+x)*np.log(1+x) + k_B*T[i]*(1-x)*np.log(1-x)

for i in range(0,len(T)):
    plt.plot(x,F(x,i),lw=1,color='r')

x=np.linspace(-0.99,0.99,num=1000)
plt.show()
```

**(v)** Minimising $F$,

$$\frac{\partial F}{\partial m}=0\implies 0=-2Jm+k_BT\ln\bigg(\frac{1+m}{1-m}\bigg)$$

and using the identity in the hint gives

$$m=\frac{k_BT}{J}\tanh^{-1}(m)$$

When the gradient of the function $k_BT\tanh^{-1}(m)/J$ at $m=0$ is greater than one, the only solution is $m=0$. If the gradient is less than one, two new solutions $m_\pm$ ($m_+=-m_-$ by symmetry) appear, as shown by the intersection of $m$ and $k_BT\tanh^{-1}(m)/J$. To evaluate this gradient, we differentiate and get

$$\frac{\partial}{\partial m}\bigg(\frac{k_BT}{J}\tanh^{-1}(m)\bigg)\bigg|_{m=0}=\frac{k_BT}{J}\frac{1}{1-m^2}\bigg\vert_{m=0}=\frac{k_BT}{J}$$

The critical temperature is therefore $T_c=\frac{J}{k_B}$.

```python
x=np.linspace(0,0.99,num=1000)
plt.plot(x,x,lw=1,color='b')
for i in range(0,len(T)):
    plt.plot(x,(T[i]/T_c)*np.arctanh(x),lw=1,color='r')

plt.show()
```

## 4. Nematic Liquid Crystal

**(i)** The vector 
$$\textbf{m}=\frac{1}{N}\sum_i\textbf{s}_i=\langle\textbf{s}\rangle$$
is a useless order paramter since $\langle\textbf{s}\rangle=\textbf{0}$. This can be reasoned from the fact that the interaction energy $E=-(\textbf{s}_i\cdot\textbf{s}_j)^2$ equally favours aligment with $\textbf{s}_i\cdot\textbf{s}_j>0$ and anti-aligment with $\textbf{s}_i\cdot\textbf{s}_j<0$. This differs from the Ising model of ferromagnetism for which aligment only is favored, hence giving non-zero $\textbf{m}$.

**(ii)** Using the summation convention,
$$S_{\alpha\alpha}=...$$

**(iii)** The trace of the matrices $\textbf{S}^2$, $\textbf{S}^3$ and $\textbf{S}^4$ may be evaluated using the summation convention and brute force multiplication.

$$\text{tr}(\textbf{S}^2)=S_{\alpha\beta}S_{\beta\alpha}=0...$$
$$\text{tr}(\textbf{S}^3)=S_{\alpha\beta}S_{\beta\alpha}=0...$$
$$\text{tr}(\textbf{S}^4)=S_{\alpha\beta}S_{\beta\alpha}=0...$$

where the result $n_\alpha n_\alpha=\textbf{n}\cdot\textbf{n}=1$ was used since $\textbf{n}$ is a unit vector.

Alternatively, by noting that the symmetric matrices $\textbf{S}^n$ ($n\geq1$) can be diagonalised, giving the eigenvalues the diagonal elements, and the trace is simply the sum of the eigenvalues.

$$\textbf{S}=\begin{pmatrix}
3Q(n_x^2-1) & 0 & 0\\
0 & 3Q(n_y^2-1) & 0\\
0 & 0 & 3Q(n_z^2-1) 
\end{pmatrix}$$

$$\textbf{S}^n=\begin{pmatrix}
3^nQ^n(n_x^2-1)^n & 0 & 0\\
0 & 3^nQ^n(n_y^2-1)^n & 0\\
0 & 0 & 3Q^n(n_z^2-1)^n 
\end{pmatrix}$$

$$\text{tr}(\textbf{S}^n)=3^nQ^n\bigg((n_x^2-1)^n+(n_y^2-1)^n+(n_z^2-1)^n\bigg)$$

The free energy is
$$f(Q)=6aQ^2+6bQ^3+18cQ^4$$
where $a,c>0$ and $b\leq0$. $f$, 
$$\frac{\partial f}{\partial Q}=0\implies 0=2aQ+3bQ^2+12cQ^3$$
which has solutions $Q=0$ and 
$$Q_\pm=\frac{-3b\pm\sqrt{9b^2-96ac}}{24c}$$ 
for $9b^2>96ac$. Since $c>0$, $f\to\infty$ as $Q\to\infty$ so $Q_+$ and $Q_-$ must be minima and maxima respectively. 

$$\frac{\partial^2f}{\partial Q^2}=12a+36bQ+216cQ^2\implies \frac{\partial^2f}{\partial Q^2}\bigg|_{Q=0}=12a>0$$

Hence $Q=0$ is always a minimum point. Initally with small negative $b$, $Q=0$ is the global minimum. As $b$ becomes more negative, the local minimum $Q_+$ develops. Beyond a critical value $b_{\text{}crit}$, $Q_+$ is a global minimum and $Q=0$ is a local minimum.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

a=1
b_crit=-np.sqrt(12*a*c)
b=np.array([0.4,0.2,0,-0.2,-0.4])
b=b+b_crit
c=1

def f(x,i):
    return 6*a*x**2 + 6*b[i]*x**3 + 18*c*x**4 

x=np.linspace(0,1.0,num=1000)

for i in range(0,len(b)):
    if i in range(0,2):
        plt.plot(x,f(x,i),lw=1,color='r')
    if i==2:
        plt.plot(x,f(x,i),lw=1,color='b')
    if i in range(3,len(b)):
        plt.plot(x,f(x,i),lw=1,color='g')
        
plt.plot(x,0*x,lw=0.5,color='k')
plt.show()
```

**(iv)** A unique non-zero solution to $f(Q)=0$ must appear at the critical point. 
$$6aQ^2+6bQ^3+18cQ^4\implies Q=\frac{-b\pm\sqrt{b^2-12ac}}{6c}$$
Hence $b^2=12ac$ at the critical point, giving
$$b_{\text{crit}}=-\sqrt{12ac}$$
and the order parameter jumps to
$$Q=-\frac{b}{6c}=\sqrt{\frac{a}{3c}}$$
Writing the order paramater more explicitly above and below the critical point

$$Q=\begin{cases}
0 & b>b_{\text{crit}}\\
\sqrt{\frac{a}{3c}} & b<b_{\text{crit}}
\end{cases}$$

which is clearly discontinous.
