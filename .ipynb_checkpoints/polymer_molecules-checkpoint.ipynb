## Isotropic Case

A polymer molecule can be modelled as $n$ small rods of length $\ell$, joined together with free hinges.

$\textbf{R}_n$ is the end-to-end vector and $\textbf{r}_n$ is the vector joining the $(n-1)^{\text{th}}$ hinge to the $n^{\text{th}}$. This effectively a random walk on $\mathbb{R}^2$ with step length $\ell$.

We would like to estimate the size of a polymer molecule in terms $n$. The average value of the end-to-end vector $\textbf{R}_n$ is a bit useless since there is no prefered direction and hence

$$\langle\textbf{R}_n\rangle=\textbf{0}~~~,$$

where the average is taken over many possible configurations of the polymer molecule. We shall instead consider the average of its square value

$$\langle\textbf{R}_n\cdot\textbf{R}_n\rangle=\langle R_n^2\rangle~~~.$$

Suppose that $\theta_n$ is the angle between the $(n-1)^{\text{th}}$ and $n^{\text{th}}$ rods. Writing $\textbf{R}_n=\textbf{R}_{n-1}+\textbf{r}_n$, we get

$$\textbf{R}_n\cdot\textbf{R}_n=\textbf{R}_{n-1}\cdot\textbf{R}_{n-1}+\textbf{r}_n\cdot\textbf{r}_n+2\textbf{R}_{n-1}\cdot\textbf{r}_n$$

$$R_n^2=R_{n-1}^2+\ell^2+2R_{n-1}\ell\cos\theta_n~~~.$$

Now taking the average, noting that $\langle\cos\theta_n\rangle=0$, a useful recursion relation is obtained:

$$\langle R_n^2\rangle=\langle R_{n-1}^2\rangle+\ell^2~~~.$$

By induction,

$$\langle R_n^2\rangle=\ell^2n$$

and therefore the root-mean-square value $\langle R_n^2\rangle^{1/2}=\ell\sqrt{n}$ provides a useful measure of polymer size.

To visualise one possible configuration of the polymer molecule, we shall take $\ell=1$ and make $n=1000$ steps starting at the origin. The end points are marked as red dots.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

ell=1
n=1000

rand.seed(1234)

R=np.array([0.0,0.0])
R2=np.array([0.0,0.0])
deltaR=np.array([0.0,0.0])

plt.plot(R[0],R[1],'ro',ms=2)

for i in range(0,n):
    theta=rand.uniform(-np.pi,np.pi)
    deltaR[0]=ell*np.cos(theta)
    deltaR[1]=ell*np.sin(theta)
    R2=R+deltaR
    plt.plot([R[0],R2[0]],[R[1],R2[1]],lw=0.2,color='b')
    R=R2

plt.plot(R[0],R[1],'ro',ms=2)    
plt.show()
```

In order to verify that $\langle R_n^2\rangle^{1/2}=\ell\sqrt{n}$, we shall simulate a large number of configurations and compute an average square distance at each step.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

ell=1
n=1000
number=100

avg=np.array([0.0])

rand.seed(123)

for i in range(0,number):

    R=np.array([0.0,0.0])
    R2=np.array([0.0,0.0])
    deltaR=np.array([0.0,0.0])

    sqD_vec=np.array([0.0])
    n_vec=np.array([0.0])

    for j in range(0,n):
        theta=rand.uniform(-np.pi,np.pi)
        deltaR[0]=ell*np.cos(theta)
        deltaR[1]=ell*np.sin(theta)
        R=R+deltaR
        sqD=R[0]**2+R[1]**2
        sqD_vec=np.append(sqD_vec,sqD)
        n_vec=np.append(n_vec,j+1)

    avg=avg+sqD_vec

avg=avg/number
avg=np.sqrt(avg)

plt.xlabel('n')
plt.ylabel('rms')
t=np.linspace(0,n,num=100)
plt.plot(t,ell*np.sqrt(t),lw=0.5,color='r')
plt.plot(n_vec,avg,lw=0.5,color='b')
plt.show()
```
