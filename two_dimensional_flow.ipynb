To conserve fluid mass, the continuity equation

$$\frac{\partial\rho}{\partial t}+\nabla\cdot(\rho\textbf{v})=0~~~,$$

where $\rho$ is density and $\textbf{v}$ is velocity must be satisfied. An incompressible fluid has constant $\rho$, so the continuity equation reduces to

$$\nabla\cdot\textbf{v}=0~~~.$$

If the fluid is furthermore irrotational, then

$$\nabla\times\textbf{v}=\textbf{0}~\implies~\textbf{v}=\nabla\phi$$

for some scalar potential $\phi$ which satisifes Laplace's equation

$$\nabla^2\phi=0~~~.$$

Consider two-dimensional flow around a sphere radius $a$. This problem lends itself to plane polar coordinates $(r,\theta)$ centred on the cylinder. The boundary conditions are

$$\frac{\partial\phi}{\partial r}\bigg\vert_{r=a}=0$$

since flow is parallel to the cylinder at its surface and 

$$\lim_{r\to\infty}\nabla\phi=\textbf{U}$$

for some constant velocity $\textbf{U}$.

Solving Laplace's equation with these boundary conditions then gives

$$\phi(r,\theta)=U\cos\theta(r+\frac{a^2}{r})~~~.$$

To plot streamlines, the streamfunction $\psi$ is first needed. By definition, it satisfies

$$\textbf{v}=\nabla\times(0,0,\psi)$$

and hence is given by

$$\psi(r,\theta)=Ur\sin\theta(1-\frac{a^2}{r^2})$$

$$\psi(x,y)=Uy(1-\frac{a^2}{x^2+y^2})~~~.$$

Streamlines are then given by the contours

$$\psi(x,y)=nc$$

where $c$ is constant and $n\in\mathbb{Z}$. Closer streamlines indicate faster flow. We must, however, be careful to only plot the physically relevant contour lines outside the cylinder.

```python
import numpy as np
import matplotlib.pyplot as plt

U=10
a=1

c=[]
integer=10
for i in range(-integer,integer):
    c.append(i)

delta=0.01
xrange=np.arange(-5,5,delta)
yrange=np.arange(-5,5,delta)
x,y=np.meshgrid(xrange,yrange)
psi=U*y*(1-a**2/(x**2+y**2))
plt.contour(x,y,psi,c,colors='b',linewidths=0.5,linestyles='solid')

plt.axis('equal')
plt.xlim(-3,3)
plt.ylim(-1,1)
plt.show()
```

To introduce some some net circulation $K$ around the cylinder, the extra term $-\frac{K}{2\pi}\log r$ is added to the streamfunction.

```python
import numpy as np
import matplotlib.pyplot as plt

U=10
a=1
K=50

c=[]
integer=10
for i in range(-integer,integer):
    c.append(i)

delta=0.01
xrange=np.arange(-5,5,delta)

yrange=np.arange(-5,5,delta)
x,y=np.meshgrid(xrange,yrange)
psi=U*y*(1-a**2/(x**2+y**2))-K*np.log(np.sqrt(x**2+y**2))/(2*np.pi)
plt.contour(x,y,psi,c,colors='b',linewidths=0.5,linestyles='solid')

plt.axis('equal')
plt.xlim(-3,3)
plt.ylim(-1,1)
plt.show()
```
